## For Single Image


```python
from PIL import Image
import numpy as np

# Point the 'test_img' to desired image file's location
test_img_loc = 'D:\\dev\\SAR\\196488331_HH.tif'
Image.MAX_IMAGE_PIXELS = None  # Override PIL's DecompressionBombError
```

```python
I = np.asarray(Image.open(test_img_loc))
```

#### Display Image

```python
from matplotlib.pyplot import imshow
%matplotlib inline

imshow(np.asarray(I))
```

### Compute Threshold

```python
threshold = np.mean(I) + 3*np.std(I)

I = I / threshold * 255
I = np.clip(I, 0, threshold)

threshold
```

### Save as npy file

```python
dump_loc = test_img_loc.replace('tif','npy')

np.save(dump_loc, I)
print(f'npy filed dumped at {dump_loc}')
```
